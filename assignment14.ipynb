Q1. Is an assignment operator like += only for show? Is it possible that it would lead to faster results at the runtime?

The assignment operator += is not just for show; it has a practical purpose. It is a shorthand for updating the value of a variable by adding a specified value to its current value. For example, a += 5 is equivalent to a = a + 5. Using the += operator can lead to faster results at runtime, especially when dealing with mutable objects like lists, as it modifies the existing object in place rather than creating a new one.

Q3. In Python, what is the most effective way to set a list of 100 integers to 0?

python
Copy code

The most effective way is to use list comprehension with the modulo operator:

Q8. Functions are said to be "first-class objects" in Python but not in most other languages, such as C++ or Java. What can you do in Python with a function (callable object) that you can't do in C or C++?

In Python, functions are first-class objects, which means you can treat them like any other object. You can do the following with functions in Python that you can't do in C or C++:

Assign functions to variables.
Pass functions as arguments to other functions.
Return functions from other functions.
Store functions in data structures like lists or dictionaries.
Q9. How do you distinguish between a wrapper, a wrapped feature, and a decorator?

In the context of programming, a wrapper is a function that provides additional functionality to another function without modifying its original code. The wrapped feature refers to the original function being wrapped by the wrapper. A decorator is a special kind of wrapper that uses the @decorator_name syntax to modify the behavior of a function or method.

Q10. If a function is a generator function, what does it return?

A generator function in Python returns a generator object. Unlike regular functions that return a value and terminate, generator functions use the yield keyword to yield a sequence of values one at a time when iterated over. The state of the function is saved between successive calls to yield, allowing efficient memory usage for large sequences.

Q11. What is the one improvement that must be made to a function in order for it to become a generator function in the Python language?

To turn a regular function into a generator function, you need to replace the return statements with yield statements. This allows the function to yield values instead of returning them, transforming it into a generator.

Q12. Identify at least one benefit of generators.

One benefit of generators is their memory efficiency. Generators produce values on-the-fly and only store the current state, rather than generating the entire sequence upfront and storing it in memory. This makes generators ideal for processing large datasets or infinite sequences without causing memory overflow.
